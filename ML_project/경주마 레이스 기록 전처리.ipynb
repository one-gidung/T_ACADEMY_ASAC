```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# one-hot encoding
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
df_horse = pd.read_csv("./data/horse_dict(경기성적).csv")
df_horse
```

```python
df_horse.info()
```

### 'Unnamed: 0' 컬럼 drop

```python
# 컬럼 버리기
df_horse.drop(columns='Unnamed: 0', inplace=True)
df_horse
```

### 군(등급) null 값인 행 삭제

```python
df_horse['군(등급)'].replace('', np.nan, inplace=True)
df_horse.dropna(subset=['군(등급)'], inplace=True)
df_horse.reset_index(inplace=True)
df_horse
```

### 군(등급) 확인

```python
df_horse.loc[:,'군(등급)'].unique()
```

```python
df_horse.loc[:,'군(등급)'].value_counts()
```

### 경주명 확인

```python
df_horse.loc[:,'경주명'].value_counts()
```

### 마체중 컬럼에서 증감 제거

```python
df_horse.loc[:,'마체중'] = df_horse.loc[:,'마체중'].apply(lambda x: x[:3])
df_horse.head()
```

### 주로 상태 라벨 인코딩
- 라벨 인코더
- 건조 : 0
- 양호 : 1
- 다습 : 2
- 포화 : 3
- 불량 : 4

```python
df_horse["주로상태"].value_counts()
```

```python
cl = {"건조    " : 0, "양호    " : 1, "다습    " : 2, "포화    " : 3, "불량    " : 4}
df_horse['주로상태'] = df_horse['주로상태'].replace(cl) # replace함수 이용 원하는 숫자로 변경
df_horse["주로상태"].value_counts()
```

### 경주일, 경주 합치기

```python
df_horse["race_id"] = ''

for i in range(len(df_horse)):
    d = df_horse.at[i, "경주일"] + '-' + str(df_horse.at[i, "경주"])
    df_horse.at[i,'race_id'] = d

df_horse
```

### 승률, 복승률 데이터 타입 변환

```python
df_horse.rename(columns = {'승률':'승률(%)','복승률':'복승률(%)'}, inplace=True)
```

```python
df_horse["승률(%)"] = df_horse["승률(%)"].str.replace(pat=r'%', repl='', regex=True)
df_horse["복승률(%)"] = df_horse["복승률(%)"].str.replace(pat=r'%', repl='', regex=True)
```

```python
df_horse["승률(%)"]= df_horse["승률(%)"].astype('float')
df_horse["복승률(%)"]= df_horse["복승률(%)"].astype('float')
df_horse.drop(columns=['index', '경주일', '경주'], inplace=True)
df_horse
```

### 기록 초수로 변경

```python
horse_time= df_horse["기록"].apply(lambda x : str(x).split(":"))
new_time = []
for i in horse_time:
  if i != ['nan']:
    second = i[1].strip()
    new_time.append(float(i[0]) * 60 + float(second))
  else :
    new_time.append(np.nan)
horse_time = pd.DataFrame(new_time, columns = ["기록"])
df_horse["기록"] = horse_time["기록"]
max_time = df_horse["기록"].max(axis=0) + 2.0
df_horse["기록"] = df_horse["기록"].fillna(max_time)
df_horse
```

```python
df_horse.info()
```

- 경기 횟수 6회 미만인 애들 삭제

```python
df_horse['race_id'].value_counts()
```

```python
vals = df_horse['race_id'].value_counts()

race_num =dict(vals)
delete_list=[]

for i, v in race_num.items():
    if v < 6:
        delete_list.append(i)

delete_list
```

```python
len(delete_list)
```

```python
asdf = df_horse
```

```python
for i in asdf.index :
    if asdf.at[i, "race_id"] in delete_list:
        asdf.drop(index = [i], inplace = True)
```

```python
asdf.reset_index(inplace=True)
asdf.drop(columns=['index'], inplace=True)
```

```python
asdf.to_csv('./data/asdf.csv')
```
