```python
# 필요없는 컬럼 제거
# 순, 데뷔일자, 비고 제거
df_jockey.drop(columns=['순', '데뷔일자', '비고'], inplace=True)
df_jockey
```

```python
# 기수명 앞에 null 지우기
df_jockey.loc[:,"기수명"] = df_jockey.loc[:,"기수명"].apply(lambda x: re.sub(r"[^ ㄱ-ㅣ가-힣+]", "", x))
df_jockey
```

```python
# 전적 컬럼 숫자 분리
temp = df_jockey.loc[:,"최근1년전적6"].apply(lambda x: re.sub(r"[^0-9]", " ", x).split())
temp
```
