```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("C:/Users/vkrwo/Documents/카카오톡 받은 파일/asdf.csv")
data.head(2)
```

```python
# 경주번호, 기록으로 정렬
data_sort = data.sort_values(by=["경주번호","기록"], ascending=[False,True])
data_sort.head(2)
```

```python
df = data_sort
df.head(2)
```

```python
final_df= pd.DataFrame()
for j in range(len(data_sort["경주번호"].unique())):
    df2 = df[df["경주번호"]==data_sort["경주번호"].unique()[j]]
    df2.index = range(len(df2))
    new_df = df2.copy()
    for i in range(len(df2)-1):
        df3 = df2.copy()
        df3 = df3.reindex( (list(range(i+1,len(df3)))) + list(range(0,i+1)) )
        df3 = df3[["마명","번호","순위","기록","수득상금","부담중량","마체중","기수","조교사"]]
        df3.columns = ["마명"+str(i+1),"번호"+str(i+1),"순위"+str(i+1),"기록"+str(i+1),"수득상금"+str(i+1),
                       "부담중량"+str(i+1),"마체중"+str(i+1),"기수"+str(i+1),"조교사"+str(i+1)] 
        df3.index = range(len(df2))
        new_df = pd.concat([new_df,df3], axis=1)
    final_df = pd.concat([final_df, new_df])
final_df
```
