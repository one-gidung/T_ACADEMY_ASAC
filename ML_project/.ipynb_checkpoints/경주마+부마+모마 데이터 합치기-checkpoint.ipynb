```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# 정규식
import re

# one-hot encoding
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

# 데이터 불러오기

```python
df_경주마 = pd.read_csv("./data/경주마데이터(최종).csv", encoding ="cp949")
df_부마 = pd.read_csv("./data/부마데이터(최종).csv", encoding ="cp949")
df_모마 = pd.read_csv("./data/모마데이터(최종).csv", encoding ="cp949")
```

```python
df_경주마
```

```python
df_부마
```

```python
# 부마성적 컬럼 숫자 분리
temp = df_부마.loc[:,"부마성적"].apply(lambda x: re.sub(r"[^0-9]", " ", x).split())
temp
```

```python
# 부마 성적 중 null값 존재
for i in range(len(temp)):
    if len(temp[i]) != 4:
        print(i, temp[i])
```

```python
# 부마 데이터에 null값 위치 확인
df_부마.iloc[121, :]
```

```python
# 널값 채우기
temp[121].append('0')
temp[121]
```

```python
# 부마성적 컬럼 분할 후 새로운 컬럼으로 넣기
df_부마["총경기_F"] = temp.apply(lambda x: int(x[0]))
df_부마["1st_F"] = temp.apply(lambda x: int(x[1]))
df_부마["2st_F"] = temp.apply(lambda x: int(x[2]))
df_부마["3st_F"] = temp.apply(lambda x: int(x[3]))

# # 기존 부마성적 컬럼 삭제
df_부마.drop("부마성적", axis=1, inplace=True)
df_부마
```

```python
df_모마
```

```python
# 모마성적 컬럼 숫자 분리
temp = df_모마.loc[:,"모마성적"].apply(lambda x: re.sub(r"[^0-9]", " ", x).split())
temp
```

```python
# 모마성적 컬럼 분할 후 새로운 컬럼으로 넣기
df_모마["총경기_M"] = temp.apply(lambda x: int(x[0]))
df_모마["1st_M"] = temp.apply(lambda x: int(x[1]))
df_모마["2st_M"] = temp.apply(lambda x: int(x[2]))
df_모마["3st_M"] = temp.apply(lambda x: int(x[3]))

# # 기존 전적 컬럼 삭제
df_모마.drop("모마성적", axis=1, inplace=True)
df_모마
```

```python
# merge를 위해 컬럼명 변경
df_부마.rename(columns={'마명':'부마명', '마번':'부마번'}, inplace=True)
df_모마.rename(columns={'마명':'모마명', '마번':'모마번'}, inplace=True)
```

```python
df_horse = pd.merge(df_경주마, df_부마, how='left', on='부마명')
df_horse = pd.merge(df_horse, df_모마, how='left', on='모마명')
df_horse
```

```python
df_horse.info()
```

```python
df_horse.drop(columns = '마번.1', inplace=True)
df_horse
```

```python
df_horse.info()
```

```python
df_horse.columns
```

```python
df_horse = df_horse.reindex(columns=['마번', '마명', '성별', '연령', '산지', 'level', '스피드(100)', '수득상금(100)', '부담력(100)', 
                                     '지구력(100)', '우승거리(100)', '레이팅', '총경기', '1st', '2st', '3st', '수득상금(천원)', 
                                     '조교사이름', '조교사번호', '마주', '부마명', '부마번', '총경기_F', '1st_F', '2st_F', '3st_F', 
                                     '모마명', '모마번', '총경기_M', '1st_M', '2st_M', '3st_M'])
df_horse.info()
```

```python
df_horse.to_csv('data/Horse_total.csv', index=False, encoding='utf-8-sig')
```
