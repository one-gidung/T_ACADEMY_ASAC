```python
a = {"기수명" : "다나카", "통산전적" : "1554전(157/141/174)"}
b = {"기수명" : "이준철", "통산전적" : "3057전(263/282/307)"}
c = {"기수명" : "우에다", "통산전적" : "272전(18/12/18)"}
d = {"기수명" : "권석원", "통산전적" : "607전(35/34/47)"}
e = {"기수명" : "다실바", "통산전적" : "4전(0/0/0)"}
df_j = df_j.append(a, ignore_index=True)
df_j = df_j.append(b, ignore_index=True)
df_j = df_j.append(c, ignore_index=True)
df_j = df_j.append(d, ignore_index=True)
df_j = df_j.append(e, ignore_index=True)
df_j
```

```python
df_j['Score'] = temp1.apply(lambda x : int(x[1])*3 + int(x[2])*2 + int(x[3])*1  / int(x[0]))
```

# 조교사 scoring

```python
# 필요없는 컬럼 제거
# 순, 데뷔일자, 비고 제거
df_t.drop(columns=['순', '소속조', '데뷔일자', '위탁관리두수', '최근1년전적'], inplace=True)
df_t
```
