# Q) id_area의 컬럼의 값을 변경
가로줄 정수 인덱스와 같은 값으로 & .대신 _로 변경  
1.North-Apulia ---------> 0_North-Apulia

```python
# lambda 함수를 이용해 변수의 모든 행의 문자열 변경
# 일시적
data["id_area"].apply(lambda x : x.replace(".", "_"))
```

```python
# x = "id_area" 변수의 각행에 들어있는 문자열
# ex) x = "1_North-Apulia"
data["id_area"] = data["id_area"].apply(lambda x : str(int(x[0])-1)+"_" + x.split(".")[1])
```

```python
# 잘 변경되었는지 확인
data.head()
```

###### id_area변수를 인덱스로 set한다

```python
data.set_index("id_area", inplace=True)
data.head()
```

```python
data.head()
```

###### palmitic 변수 100으로 나눠주기

```python
data["palmitic"] = data["palmitic"].apply(lambda x : x/100)
data.head()
```
