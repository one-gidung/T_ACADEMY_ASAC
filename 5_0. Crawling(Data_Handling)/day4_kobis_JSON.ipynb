```python
# 잠시 초기화 이상한 코드를...
movie_df = pd.DataFrame()
movie_df["movieCd"] = ""
movie_df["movieNm"] = ""
movie_df["Director"] = ""
movie_df
```

```python
# 일단 기본적인 1개씩 축적하는 방식
# 1개 정보를 수집할 임시 DF
# 여러 개 대한 정보를 쌓을  DF
movie_df = pd.DataFrame()
movie_df_temp = pd.DataFrame(
    data = {
    "movieCd" : "", 
    "movieTitle" : "", 
    "movieETitle" : "", 
    "openDay" : "", 
    "DirName" : "",
    "movieType" : ""},
    index=["i-data"]
)

# 본인이 만든 컬럼이름이 몇 번째 위치인지 정수 인덱스
# json의 값이 아니라 본인이 만든 DF의 컬럼이름
temp_code_id = movie_df_temp.columns.get_loc("movieCd")
temp_title_id = movie_df_temp.columns.get_loc("movieTitle")
temp_etitle_id = movie_df_temp.columns.get_loc("movieETitle")
temp_open_id = movie_df_temp.columns.get_loc("openDay")
temp_dir_id = movie_df_temp.columns.get_loc("DirName")

temp_type_id = movie_df_temp.columns.get_loc("movieType")

movie_df_temp
```

```python
# 데이터를 합칠 때 concat을 사용 ---> pd.concat(["합칠것들 나열1", "합칠것들22"])
# 업데이트 하려면 concat한 것을 변수에 재할당 해줘야 한다
# 합친것은 저장되지 않고 당장의 출력값만 보여주기 때문
pd.concat([movie_df, movie_df_temp], ignore_index=True)
```

```python
# 처리해야할 데이터의 수 = 50
num = len(movie_data["movieListResult"]["movieList"])
for i in range(num):
    # i 번째 영화의 정보를 담고
    temp = movie_data["movieListResult"]["movieList"][i]
    
    # i 번째 영화의 상세 정보들을 movie_df_temp에 담기
    movie_df_temp.iat[0,temp_code_id] = temp["movieCd"]
    movie_df_temp.iat[0,temp_title_id] = temp["movieNm"]
    movie_df_temp.iat[0,temp_etitle_id] = temp["movieNmEn"]
    movie_df_temp.iat[0,temp_open_id] = temp["openDt"]
    movie_df_temp.iat[0,temp_type_id] = temp["typeNm"]
    # 감독 상세 정보
    if temp["directors"] == []:
        movie_df_temp.iat[0,temp_dir_id] = ""
    else:
        movie_df_temp.iat[0,temp_dir_id] = temp["directors"][0]["peopleNm"]
    
    # 축적하는 곳으로 가서 쌓기
    movie_df = pd.concat([movie_df, movie_df_temp], ignore_index=True)
    
    # temp에 대해서 비우기 ~~ i+1번째 영화 정보 담을 준비
    for j in range(len(movie_df_temp.columns)):
        movie_df_temp.iat[0, j] = ""
    
movie_df
```

```python
# 1) 위의 코드를 컬럼 중심으로 처리해보세요 --> get_loc etc
# 2) 위의 값 이외의 다른 정보들도 추가 수집해서 DF를 완성
#    ** 필수항목 : 코드, 제목, 개봉일, 감독1번 이름
# 3) 2번의 개봉일 중심으로
#    year, month, day 컬럼을 추가할 것 (기존 openDT컬럽 apply, lambda)
#    기존 개봉날자 컬럼을 기반으로 apply, lambda를 써서
#    새로운 year, month, day 컬럽을 생성
# 4) movieCd 컬럼(코드값)을 가로줄에 대한 인덱스로 지정
#    참고) set_index
# 5) 개봉날짜별로 정렳해서 csv/excel 파일로 출력
```

```python
# movieCd가 유니크한 값인지
# 인덱스로 지정해도 될런지 확인
# True!!
len(movie_df["movieCd"].unique()) == len(movie_df)
```

```python
# columns: 코드, 제목, 영어제목, 개봉일, year, month, day, 감독1번이름
movie_df["year"] = movie_df["openDay"].apply(lambda x : x[:4])
movie_df["month"] = movie_df["openDay"].apply(lambda x : x[4:6])
movie_df["day"] = movie_df["openDay"].apply(lambda x : x[6:])

# 개봉날짜별로 정렬

movie_df.sort_values(by = "openDay", ascending = False, inplace=True)


# movieCd 인덱스 지정
movie_df.set_index("movieCd", drop=True, inplace=True)

movie_df
```

내가 원하는 것은 for문을 돌면서 딱 특정한 지점 1개에 삽입하고 싶기 때문에  
at, iat을 사용해야한다  
위에 내가 짠 것처럼 loc으로 하면 Series를 받기 때문에 Series에서 한 번더 index를 search해야하기 때문에  
빨간색 경고창이 뜬 것이다.  
낭비이다!!!!!!  
아래와 같은 코드가 훨 좋다.

```python
# 내 코드를 loc --> at 으로 수정한 버전
url_base = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json"
key = "edef7c61e5650b860807260fc96f9504"

for i in movie_df.index:
    # i번째 영화에 대해서
    url = url_base + '?key=' + key + '&movieCd=' + i
    res = urllib.request.urlopen(url)
    temp_data = json.loads(res.read().decode('utf-8'))
    # 필요한 데이터는 genre, actnum, actname, actcast
    movie_info = temp_data['movieInfoResult']['movieInfo']
    # genre
    movie_df.at[i, 'Genre'] = movie_info['genres'][0]['genreNm']
    movie_df.at[i, 'ActNum'] = str(len(movie_info['actors']))
    if len(movie_info['actors']) != 0:
        movie_df.at[i, 'ActName'] = movie_info['actors'][0]['peopleNm']
        if movie_info['actors'][0]['cast'] == '':
            movie_df.at[i, 'ActCast'] = None
        else:
            movie_df.at[i, 'ActCast'] = movie_info['actors'][0]['cast']
    else:
        movie_df.at[i, 'ActName'] = None
        movie_df.at[i, 'ActCast'] = None
movie_df
```
