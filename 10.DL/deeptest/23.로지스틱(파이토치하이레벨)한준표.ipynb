```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch
from torch.optim import Adam
from torch.nn import Linear, BCELoss, Sequential, Sigmoid
```

```python
x_data = np.array([[1,3],[2,2],[3,1],[4,6],[5,5],[6,4]])
y_data = np.array([[0],[0],[0],[1],[1],[1]])
```

```python
x = torch.FloatTensor(x_data)
y = torch.FloatTensor(y_data)
```

```python
model = Sequential()
model.add_module("nn1", Linear(2,1)) # w 2x1 // b 1
model.add_module("sig1", Sigmoid()) # activation function
loss_fn = BCELoss()
optimizer = Adam(model.parameters(), lr = 0.1)
```

```python
cost_hist=[]
for epochs in range(1000):
    hx = model.forward(x) 
    optimizer.zero_grad()
    c = loss_fn(hx, y)
    c.backward()
    cost_hist.append(c)
    optimizer.step()
    print("cost : ", c.item(), "/ acc : ", accuracy_score(torch.IntTensor(y_data),(model.forward(x).detach().numpy()>0.5)+0))
```
