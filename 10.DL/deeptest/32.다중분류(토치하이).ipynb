```python
import torch
from torch.optim import Adam
from torch.nn import Linear, CrossEntropyLoss, Sequential, Softmax

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv( 'data/softmax.txt', sep=' ',header=None)
df
```

```python
x_data = df.iloc[:,:-3].values
y_data = df.iloc[:,2:].values
```

```python
x = torch.FloatTensor( x_data )
y = torch.FloatTensor( y_data )
```

```python
model = Sequential()
model.add_module( 'nn1', Linear(2,3) ) # w:2x3, b:3 
model.add_module( 'soft1', Softmax(dim=1) ) #활성함수.
loss_fn = CrossEntropyLoss()
optimizer = Adam( model.parameters(), lr=0.1 )
```

```python
for epoch in range(2000):
    optimizer.zero_grad()
    hx = model(x) # model.forward(x)  z = matmul( x,w) +b ==> Softmax( z , dim=1)
    c = loss_fn( hx, y) 
    c.backward()
    optimizer.step()
    print( epoch, c.item() )
```
