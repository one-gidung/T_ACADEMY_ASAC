```python
import tensorflow as tf
from tensorflow.keras.optimizers import Adam
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv( 'data/test.csv', header=None)
df.columns = ['q1','q2','mid','final']
df
```

```python
x_data = df.iloc[:, :-1]
y_data = df.iloc[:, [-1] ]
```

```python
x_data
```

```python
x_data.shape
```

```python
x = tf.constant( x_data, tf.float32)
y = tf.constant( y_data, tf.float32)
```

```python
w = tf.Variable( tf.random.uniform([3,1]) )#[특성데이터갯수, 라벨의갯수]
b = tf.Variable( tf.random.uniform([1]) ) # 라벨의 갯수
```

```python
def cost():
#     hx = x1*w1 + x2*w2 + x3*w3 +  b
    hx = tf.matmul( x, w ) + b
    c = tf.reduce_mean( (hx-y)**2)
    return c
```

```python
optimizer = Adam( learning_rate=0.1 )
hist_cost=[]
for i in range( 1000):
    optimizer.minimize( cost, var_list=[w,b] )
    c = cost().numpy()
    hist_cost.append( c )
    print( i, 'cost=', c )
```
