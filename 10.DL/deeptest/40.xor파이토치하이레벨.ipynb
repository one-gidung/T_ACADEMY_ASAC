```python
import torch
from torch.optim import Adam
from torch.nn import Linear, MSELoss, Sequential,BCELoss, Sigmoid

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
x_data = np.array( [[0,0],[0,1],[1,0],[1,1]] )
y_data = np.array( [[0],[1],[1],[0] ] )
```

```python
x = torch.FloatTensor( x_data )
y = torch.FloatTensor( y_data )
```

```python
model = Sequential()
model.add_module( 'nn1', Linear(2,32) ) # w:2x32, b:32 
model.add_module( 'sig1', Sigmoid() ) #활성함수.
model.add_module( 'nn2', Linear(32,1) ) # w:32x1, b:1 
model.add_module( 'sig2', Sigmoid() ) #활성함수.

loss_fn = BCELoss()
optimizer = Adam( model.parameters(), lr=0.1 )
```

```python
for epoch in range(2000):
    optimizer.zero_grad()
    hx = model(x) 
    # model.forward(x)  z1 = matmul( x,w1) +b1 ==> Sigmoid( z1 ) => z=matmul(h1,x2) + 
    c = loss_fn( hx, y) 
    c.backward()
    optimizer.step()
    acc = accuracy_score( y_data, ((hx>0.5)+0 ).detach().numpy() )
    print( epoch, c.item(), acc )
```
