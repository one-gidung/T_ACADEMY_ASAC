```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.optimizers import Adam
```

```python
df = pd.read_csv( 'data/softmax.txt', sep=' ',header=None)
df
```

```python
x_data = df.iloc[:,:-3]
y_data = df.iloc[:,2:]
```

```python
x_data.shape
```

```python
y_data.shape
```

```python
x = tf.constant( x_data, tf.float32)
y = tf.constant( y_data, tf.float32)
```

```python
initX =tf.initializers.GlorotUniform()
w = tf.Variable( initX( shape=[2,3] ) ) # 특성데이터갯수, 라벨의 갯수
b = tf.Variable( initX( shape=[3] ) ) #라벨의갯수(분류의갯수)
```

```python
def cost():
    z = tf.matmul(x,w) + b
#     hx = tf.nn.softmax( z )
    c_i = tf.nn.softmax_cross_entropy_with_logits( labels=y, logits=z)
    c = tf.reduce_mean( c_i )
    return c
```

```python
optimizer = Adam(0.1)
for i in range(1000):
    optimizer.minimize( cost, var_list=[w,b] )
    print(i, 'cost:', cost().numpy() )
```

```python
def hxFn( xdata ):
    xd = tf.constant( xdata, tf.float32)
    z = tf.matmul( xd, w ) + b
    hx = tf.nn.softmax( z )
    return hx.numpy().argmax(axis=1)
#     return z.numpy(), hx.numpy()
```
