```python
import tensorflow as tf
import numpy as np
import pandas as pd

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
```

```python
x_data = np.array( [[0,0],[0,1],[1,0],[1,1]] )
y_data = np.array( [[0],[1],[1],[0] ] )
```

```python
dense1 = Dense( units=100, input_dim=2 , activation='sigmoid' ) #w:2x100, b: 100
dense2 = Dense( units=1, input_dim=100 , activation='sigmoid' ) #w:100x1 b:1
model = Sequential( [dense1, dense2] )
model.compile( loss='binary_crossentropy', optimizer=Adam(0.1), metrics=['acc']  )
h = model.fit( x_data, y_data, epochs=1000)
```

```python
model.predict( x_data ) # 4x2 2x100 => 4x100 100x1 4x1
```

```python
(model.predict( x_data )>0.5) +0
```

```python
modelNN = Sequential()
modelNN.add( Dense( units=100, input_dim=2 , activation='relu' ) )
modelNN.add( Dense( units=1, input_dim=100 , activation='sigmoid' ) )
modelNN.compile( loss='binary_crossentropy', optimizer=Adam(0.1), metrics=['acc']  )
h = modelNN.fit( x_data, y_data, epochs=1000)
```
