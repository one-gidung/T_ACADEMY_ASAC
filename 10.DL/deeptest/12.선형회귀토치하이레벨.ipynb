```python
import torch
from torch.optim import Adam
from torch.nn import Linear, MSELoss, Sequential

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/cars.csv', index_col='Unnamed: 0')
df
```

```python
df[['speed']].values.shape
```

```python
x= torch.FloatTensor( df[['speed']].values )
y= torch.FloatTensor( df[['dist']].values )
```

```python
linear = Linear( 1,1 ) # 특성데이터갯수, 라벨의 갯수 w,b 
```

```python
linear.weight
```

```python
linear.bias
```

```python
list( linear.parameters() )
```

```python
model = Sequential()
model.add_module( 'nn1', linear )
loss_fn = MSELoss()
optimizer = Adam( model.parameters(), lr= 0.1 )
```

```python
for step in range(1000):
#     hx = linear.forward(x) # w*x + b
    hx = model.forward(x)
    # hx = linear(x)  #linear.forward(x)
    optimizer.zero_grad()
    c = loss_fn( hx, y )
    c.backward()
    optimizer.step()
    print( 'cost', c.item() )
```
