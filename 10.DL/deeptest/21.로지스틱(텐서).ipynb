```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.optimizers import Adam
```

```python
x_data = np.array( [[1,3],[2,2],[3,1],[4,6],[5,5],[6,4]]) # 공부한시간,출석일수
y_data = np.array( [[0],[0],[0],[1],[1],[1]])
```

```python
x_data.shape
```

```python
x = tf.constant( x_data, tf.float32)
y = tf.constant( y_data, tf.float32)
```

```python
# xavier, he( 최적의 초기화 알고리즘)
initX =tf.initializers.GlorotUniform()
w = tf.Variable( initX( shape=[2,1] ) )
b = tf.Variable( initX( shape=[1] ) )
```

```python
def cost():
    z = tf.matmul(x,w) + b
    #     hx = tf.nn.sigmoid(z)
    c_i = tf.nn.sigmoid_cross_entropy_with_logits( labels=y, logits=z)
    c = tf.reduce_mean( c_i )
    return c
```

```python
optimizer = Adam(0.1)
for i in range(1000):
    optimizer.minimize( cost, var_list=[w,b] )
    print(i, 'cost:', cost().numpy() )
```

```python
def hxFn( xdata ):
    xd = tf.constant( xdata, tf.float32)
    z = tf.matmul( xd, w ) + b
    hx = tf.nn.sigmoid( z )
    return (hx.numpy()>0.5) + 0
#     return z.numpy(), hx.numpy()
```

```python
hxFn( [[5,5]] )
```

```python
hxFn( [[1,1]] )
```

```python
pred = hxFn( x_data )
pred
```

```python
(pred == y_data).mean()
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score( y_data, pred )
```
