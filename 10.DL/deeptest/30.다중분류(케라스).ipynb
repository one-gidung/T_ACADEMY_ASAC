```python
import tensorflow as tf
import numpy as np
import pandas as pd

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv( 'data/softmax.txt', sep=' ',header=None)
df
```

```python
x_data = df.iloc[:,:-3]
y_data = df.iloc[:,2:]
```

```python
dense = Dense( units=3, input_dim=2 , activation='softmax' ) # w:2x3 b:3
model = Sequential( [dense] )
model.compile( loss='categorical_crossentropy',
              optimizer=Adam(0.1), metrics=['acc']  )
h = model.fit( x_data, y_data, epochs=1000)
```

```python
w, b = dense.get_weights()
```

```python
model.predict( [[1,1],[5,6]] ).argmax(axis=1)
```

```python
model.evaluate( x_data, y_data)
```
