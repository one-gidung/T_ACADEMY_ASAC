$ \frac{dy}{dx} = 3*2x^2 + 0 $  
$ y^\prime = 6x^2 $  

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("./data/electric.csv", index_col=0)
df
```

```python
x  =  np.array( df['전기사용량'] )
y = np.array( df['전기생산량'] )
```

```python
plt.xlim( 0,6 )
plt.ylim( 0,6 )
plt.scatter( x, y)
plt.show()
```

```python
w = 10
b = 10
epochs = 2000
learning_rate =0.01
n = len(x)
c_hist=[]
for i in range( epochs):
    hx = w*x + b
    cost = np.sum( (hx-y)**2 )/n
    c_hist.append( cost )
    gradientW = np.sum( ( w*x + b-y )*2*x )/n
    gradientB = np.sum( ( w*x + b-y )*2 )/n
    w = w-learning_rate*gradientW
    b = b-learning_rate*gradientB
    print( 'cost', cost, 'w:',w, 'b:', b )
```

```python
plt.plot( c_hist )
plt.show()
```

```python
print('최종w', w, '최종b', b)
```
