```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_data = np.array( [1,2,3] )
y_data = np.array( [1,2,3] )
```

```python
def cost( x,y,w):
    hx = w*x # [-1,-2,-3]
    c = np.sum( (hx-y)**2 ) # [-1,-2,-3] - [1,2,3] = [-2,-4,-6]= [4,16,36]
    return c/len(x)
```

```python
def gradient( x,y,w):
    c = np.sum( (w*x-y)*2*x  )
    return c/len(x)
```

```python
print( cost(x_data,y_data,-1) )
print( cost(x_data,y_data,0) )
print( cost(x_data,y_data,1) )
print( cost(x_data,y_data,2) )
print( cost(x_data,y_data,3) )
```

```python
w= 10
for epoch in range(200):
    c = cost( x_data, y_data, w)
    g = gradient( x_data, y_data, w)
    w = w-0.1*g
    print( 'epoch', epoch, 'cost=',c, 'w=',w )
```
