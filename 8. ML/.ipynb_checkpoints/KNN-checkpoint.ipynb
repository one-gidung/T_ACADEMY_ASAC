```python
import numpy as np
import pandas as pd
```

```python
# ML을 활용하기 위해서 scikit-learn을 부른다.
# 평가)
from sklearn import metrics

# 데이터를 학습용과 평가용으로 나눠주는
from sklearn.model_selection import train_test_split
```

```python
# toy-example Dataset: MNIST
from sklearn.datasets import fetch_openml
```

```python
# 데이터 불러오기
mnist = fetch_openml('mnist_784', version=1, as_frame=False)
mnist
```

```python
# 데이터 분리: train_test_split
# 단, 지금은 array 형식에서 한다. (보통은 DF)
# 모의문제, 실전문제, 모의정답, 실제정담 순으로 넣어줘야함
(train_data, test_data, train_label, test_label) = train_test_split(
    np.array(mnist.data), # 전체 문제
    mnist.target,         # 전체 문제에 대한 정답
    test_size = 0.25,
    random_state = 1234
)
```

```python
# 일단 데이터를 수집을 했다고 가정하고....
# 정규화 작업( 숫자들의 값을 조정!!!! )
# 대표적 : minMaxScaler(min=0, max=1), 
#          standScaler(평균, 표준편차) Z = x-m/sigma
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# A : 10, 20, 30 --> min10, max 30 --> 0, 0.5 ,1
# B : 100, 200, 300 --> min100, max300 --> 0, 0.5, 1
# C : 1,  20, 100 --> min1, max100 ---> 0, 0.2 ,1
```

```python
# 모델처럼 생각을 함.
min_max_sc = MinMaxScaler()

# 주어진 데이터를 기반으로 룰을 잡고,그 룰대로 값을 변환
# fit_transform , fit, transform
train_norm = min_max_sc.fit_transform(train_data)
test_norm = min_max_sc.transform(test_data)

# 이렇게 해도 늘 뭐를 신경써야 하는가하면....
# test에서 unseen value, data가 발생하면...에러가 뜰 수 있음
# test는 그냥 나 몰라요 생각하고,
# 내가 원하는대로 할 수 있는 것은 train만 있다고 생각하면 됨
# 그런데, train에서 보지못한 것들이 test에서 나타날 수 있다는 것을 염두해두어야 한다.
```

```python
from sklearn.neighbors import KNeighborsClassifier
# f(x)의 모양을 결정!!! --> parameter~~~~~~
# parameter들을 결정해주는게 fit
```

```python
# 평가
from sklearn import metrics # -> metrics.accuracy_score
from sklearn.metrics import accuracy_score # accuracy_score
# 시간
import time
```

n_jobs는 cpu core에 관한 것이다...  

```python
t1 = time.time()
# 1) 내가 사용할 모델을 볼러온다
knn = KNeighborsClassifier(n_neighbors=9, n_jobs=-1)
# 2) 학습 : 1번에서 세팅한 모델의 최적의 파라미터를 내 데이터에 fit 메서드 한다.
knn.fit(train_data, train_label)    # 시간이 오래 걸리는 부분
# 3) 2에서 만든 모델을 가지고 test_data 실전문제 풀기
pred = knn.predict(test_data)
# 4) test data가 train data로 훈련한 모델에 얼마나 잘 맞는지 확인한다
# 나는 여기서 accuracy를 사용한다.
print(accuracy_score(test_label, pred))
print("수행시간:", str(time.time()-t1))
```
