```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.optimizers import Adam
import pandas as pd

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
df = pd.read_csv('data/pima-indians-diabetes.data.csv')
df
```

```python
x_data = df.iloc[:,:-1].values
y_data = df.iloc[:,[-1]].values
```

```python
def createModel():
    dense = Dense( units=1, input_dim=8, activation='sigmoid')
    model = Sequential( [dense] )
    model.compile( loss='binary_crossentropy', optimizer=Adam(0.1),
                 metrics=['acc'])
    return model
```

```python
kcf = KerasClassifier( build_fn=createModel, epochs=1000, verbose=1 )
```

```python
pmodel = make_pipeline( MinMaxScaler(), kcf )
pmodel.fit( x_data, y_data )
```

```python
pmodel.predict( [[6,148,72,35,0,33.6,0.627,50]] )
```
