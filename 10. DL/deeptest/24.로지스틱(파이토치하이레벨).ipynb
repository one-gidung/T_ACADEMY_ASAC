```python
import torch
from torch.optim import Adam
from torch.nn import Linear, MSELoss, Sequential,BCELoss, Sigmoid

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
x_data = np.array( [[1,3],[2,2],[3,1],[4,6],[5,5],[6,4]])
y_data = np.array( [[0],[0],[0],[1],[1],[1]])
```

```python
x = torch.FloatTensor( x_data )
y = torch.FloatTensor( y_data )
```

```python
model = Sequential()
model.add_module( 'nn1', Linear(2,1) ) # w:2x1, b:1 
model.add_module( 'sig1', Sigmoid() ) #활성함수.
loss_fn = BCELoss()
optimizer = Adam( model.parameters(), lr=0.1 )
```

```python
for epoch in range(1000):
    optimizer.zero_grad()
    hx = model(x) # model.forward(x)  z = matmul( x,w) +b ==> Sigmoid( z )
    c = loss_fn( hx, y) 
    c.backward()
    optimizer.step()
    acc = accuracy_score( y_data, ((hx>0.5)+0 ).detach().numpy() )
    print( epoch, c.item(), acc )
```
