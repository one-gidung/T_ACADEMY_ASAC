```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential, Input, Model
from tensorflow.keras.optimizers import Adam
```

```python
x_data = np.array( [[0,0],[0,1],[1,0],[1,1]] )
y_data = np.array( [[0],[1],[1],[0] ] )
```

```python
dense1 = Dense( units=100, input_dim=2, activation='sigmoid')
dense2 = Dense( units=1, activation='sigmoid')
model1 = Sequential( [dense1, dense2])
model1.compile( loss='binary_crossentropy', optimizer='adam', metrics=['acc'] )
model1.fit( x_data, y_data, epochs=500)
```

```python
model2 = Sequential()
model2.add( Dense( units=100 ,input_dim=2, activation='sigmoid') )
model2.add( Dense( units=1, activation='sigmoid') )
model2.compile( loss='binary_crossentropy', optimizer='adam', metrics=['acc'] )
model2.fit( x_data, y_data, epochs=500)
```

```python
myin = Input(shape=(2,) )
out1 = Dense( units=100, activation='relu')( myin )
out2 = Dense( units=1, activation='sigmoid')( out1 )
model3 = Model( myin, out2 )
model3.compile( loss='binary_crossentropy', optimizer='adam', metrics=['acc'] )
model3.fit( x_data, y_data, epochs=500)
```
