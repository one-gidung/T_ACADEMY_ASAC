```python
class XORModel( Module ):
    def __init__(self):
        super().__init__() #상위 생성자 호출
        self.nn1 = Linear( 2,32) # w1: 2x32  b:32
        self.sig1 = Sigmoid()
        self.nn2 = Linear( 32,1) # w2:32x1 b:1
        self.sig2 = Sigmoid()
    
    def forward( self, x):
        print( 'forward call')
        x = self.nn1(x) # self.nn1.forward(x)  matmul( w1,x)
        x = self.sig1(x) # self.sig1.forward( x ) 1/1+e-z
        x = self.nn2(x)
        x = self.sig2(x)
        return x
```

```python
x_data = np.array( [[0,0],[0,1],[1,0],[1,1]] )
y_data = np.array( [[0],[1],[1],[0] ] )
x = torch.FloatTensor( x_data )
y = torch.FloatTensor( y_data )
```

```python
model = XORModel()
loss_fn = BCELoss()
optimizer = Adam( model.parameters(), lr=0.1)
```

```python
for epoch in range(2000):
    optimizer.zero_grad()
    hx = model(x) #model.forward(x)
    cost = loss_fn( hx, y )
    cost.backward()
    optimizer.step()
    print( 'epoch:', epoch, 'cost:', cost.item() )
```
