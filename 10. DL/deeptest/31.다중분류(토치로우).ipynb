```python
import torch
from torch.optim import Adam
import torch.nn.functional as F
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv( 'data/softmax.txt', sep=' ',header=None)
df
```

```python
x_data = df.iloc[:,:-3].values
y_data = df.iloc[:,2:].values
```

```python
x = torch.FloatTensor( x_data )
y = torch.FloatTensor( y_data )
```

```python
w = torch.empty( [2,3],requires_grad=True) # [특성데이터갯수, 라벨갯수]
b = torch.empty( [3],requires_grad=True)# [라벨갯수]
torch.nn.init.uniform_(w)
torch.nn.init.uniform_(b)
```

```python
def cost():
    z =torch.matmul(x,w) + b
#     hx = F.softmax( z ,dim=1)
#     cost_i = (y*-torch.log( hx)).sum()    
    c_i = F.cross_entropy(z, y)
    c = torch.mean( c_i )
    return c
```

```python
optimizer = Adam( [w,b] ,lr=0.01 )
for epoch in range(2000):
    c = cost()
    optimizer.zero_grad()
    c.backward()
    optimizer.step()
    print('cost', c.item() )
```

```python
def hxFn( xdata ):
    xd = torch.FloatTensor( xdata )
    z = torch.matmul( xd,w) + b
    hx = torch.softmax( z , dim=1)
    return hx.argmax( dim=1).detach().numpy() # torch.argmax(hx, dim=1)
#     return z, hx
#     return hx.detach().numpy()
```
