```python
import tensorflow as tf
from tensorflow.keras.optimizers import Adam, SGD
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
data = [[828, 920, 1234567, 1020, 1111],
            [824, 910, 2345612, 1090, 1234],
            [880, 900, 3456123, 1010, 1000],
            [870, 990, 2312123, 1001, 1122],
            [860, 980, 3223123, 1008, 1133],
            [850, 970, 2432123, 1100, 1221]]
data = np.float32(data)
```

```python
df = pd.DataFrame(data)
df
```

```python
x_data = df.iloc[:,:-1]
y_data = df.iloc[:, [-1] ]
```

```python
scaleF = MinMaxScaler()
x_dataN = scaleF.fit_transform( x_data)
```

```python
scaleL = MinMaxScaler()
y_dataN = scaleL.fit_transform( y_data)
```

```python
x_dataN
```

```python
y_dataN
```

```python
x = tf.constant( x_dataN, tf.float32)
y = tf.constant( y_dataN, tf.float32)
```

```python
w = tf.Variable( tf.random.uniform([4,1]) )#[특성데이터갯수, 라벨의갯수]
b = tf.Variable( tf.random.uniform([1]) ) # 라벨의 갯수
```

```python
def cost():
    hx = tf.matmul( x, w ) + b
    c = tf.reduce_mean( (hx-y)**2)
    return c
```

```python
optimizer = Adam( learning_rate=0.1 )
hist_cost=[]
for i in range( 1000):
    optimizer.minimize( cost, var_list=[w,b] )
    c = cost().numpy()
    hist_cost.append( c )
    print( i, 'cost=', c )
```

```python
def hxFn( xdata ):
    xd = tf.constant( xdata, tf.float32)
    hx = tf.matmul( xd, w ) + b
    return hx.numpy()
```

```python
xn = scaleF.transform( [[828, 920, 1234567, 1020]] )
```

```python
pred = hxFn( xn )
```

```python
scaleL.inverse_transform( pred )
```
