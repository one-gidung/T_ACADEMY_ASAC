### Hannanum
Hannanum을 이용해 명사, 형태소, 품사를 추출하시오.

```python
from konlpy.tag import Hannanum
```

```python
han = Hannanum()
```

```python
# 명사 추출
han.nouns(c)
```

```python
# 형태소 추출
han.morphs(c)
```

```python
# 품사 추출
han.pos(c)
```

### Okt
okt를 이용해서 헌법에서 형용사만 추출하시오.

```python
from konlpy.tag import Okt
```

```python
okt = Okt()
```

```python
# 방법 1
# 형용사만 추출
a = okt.pos(c)
for key, value in a:
    if value == 'Adjective':
        print(key)
```

```python
# 방법 2
# 형용사만 추출 다른 방법
[n[0] for n in okt.pos(c) if n[1]=='Adjective']
```

```python
from nltk import Text

import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
txtLaw = Text( han.nouns( c ))
txtLaw.plot(30)
plt.show()
```

```python
txtLaw = Text( okt.nouns( c ))
txtLaw.plot(30)
plt.show()
```

```python
result = [ n for n in okt.nouns( c ) if len(n)>=2 ]
txtLaw = Text( result )
txtLaw.plot(30)
plt.show()
```

### 그림 양 옆으로 이쁘게 그려보기 (혼자 그냥 해본 거), hannanum과 okt 그림 비교

```python
import pandas as pd
```

```python
txtLaw1 = Text( han.nouns( c ))
txtLaw2 = Text( okt.nouns( c ))

mydf1 = pd.DataFrame( txtLaw1.vocab().items() )
mydf2 = pd.DataFrame( txtLaw2.vocab().items() )

mydf1 = mydf1.sort_values(1, ascending=False)
mydf2 = mydf2.sort_values(1, ascending=False)
```

```python
fig = plt.figure(figsize=(20, 8))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.plot(mydf1[0][:30], mydf1[1][:30])
ax2.plot(mydf2[0][:30], mydf2[1][:30])
plt.show()
```
