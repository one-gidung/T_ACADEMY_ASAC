```python
import pandas as pd
import requests
import time
import xmltodict
```

### Data parsing

```python
url = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=109"

response = requests.get(url)
time.sleep(3)
data_json = xmltodict.parse(response.text)
```

```python
data_json
```

- 필요한 데이터가 있는 쪽으로 keys()를 이용해 파고들어가서 list_df에 저장

```python
list_df = data_json['rss']['channel']['item']['description']['body']['location']
list_df
```

```python
#지역 날짜 날씨 최고 최저 기온
# 지역: city
# 날짜: data -> tmEf
# 날씨: data -> wf
# 최고기온: data -> tmx
# 최저기온: data -> tmn
```

```python
total_list = list()

for i in range(len(list_df)):
    my_list = list()
    for j in range(len(list_df[i]['data'])):
        my_list.append(pd.DataFrame(list_df[i]['data'][j], index=[j]))
    temp = pd.concat(my_list)
    temp['city'] = list_df[i]['city']
    total_list.append(temp)
total_list
```
