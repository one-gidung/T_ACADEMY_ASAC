```python
#### DF이 정해지면, 내가 원하는 형태로 테이블을 가로, 세로 만들겠다는 것!!!
#### 엑셀의 pivot 기능을 하고자 함!!!!!!
```

```python
import pandas as pd
import numpy as np
```

```python
path = './data/11_sales-funnel.xlsx'
data = pd.read_excel(path, sheet_name = "Sheet1")

data.head()
```

```python
data.shape
```

```python
# 영업 관련 데이터입니다!!!!!! --> 영업팀에 인사 평가를 진행하려고 함!!!!
# 본인은 인사팀의 팀장이라 생각하고 이 데이터를 보시면 됩니다!!!!!!!
# 고객 계좌번호, 고객이름, 영업사원, 영업사원팀장, 상품, 수량, 가격, 진행상태
```

```python
# pivot에 대해서 내가 보고자하는 가로, 내가 보고자하는 세로!!!!!
```

```python
# 1) 가로줄에 대한 세팅 : index 파라미터로 지정!!!!!!
```

```python
# 1-1) 내가 보고싶은 것이 고객이름별로 항목들을 보자!!!
```

```python
pd.pivot_table(data, index=["Name"])
```

```python
# 고객하고 그 고객 영업담당자
pd.pivot_table(data, index=["Name", "Rep"])
# 18  by 7 ---> 13 
```

```python
# 인사팀장의 입장에서는 영업사원과 담당팀장
# 팀장2명이 잇는데, 누굴 승진 시켜야 하나?
pd.pivot_table(data, index=["Manager", "Rep"])
```

```python
# 2) 세로줄에서 내가 보고자 하는 항목을 선택 : values
```

```python
# 2-1) 세로줄에서 price만 보고자 합니다...
pd.pivot_table(data, index=["Manager", "Rep"], 
              values=["Price"])
```

```python
# 3) 내가 가로/세로로 만든 테이블에서 무엇으로 채울까.... aggfunc
pd.pivot_table(data, index=["Manager", "Rep"], 
              values=["Price"],
              aggfunc=[np.sum])
# 참고) aggfunc : 함수명만 ()를 빼고 함수명만 사용
```

```python
pd.pivot_table(data, index=["Manager", "Rep"], 
              values=["Price"],
              aggfunc=[np.sum, len])
```

```python
# 4) 항목별로 나눠서 볼 때 : columns
```

```python
# 4-1) 3번의 매출실적은 알겠는데,,,,어떤 상품을 잘 팔았는지??
pd.pivot_table(data, index=["Manager","Rep"], 
               values=["Price"],
               columns=["Product"],
               aggfunc=[np.sum])
```

```python
# 채울 수 없는 공간이 생길 수 있어요!!!!!!   ---->  빵구난 거 채우는 거 fillna

pd.pivot_table(data, index=["Manager","Rep"], 
               values=["Price"],
               columns=["Product"],
               aggfunc=[np.sum], fill_value=0)
```

```python
pd.pivot_table(data, index=["Manager","Rep"], 
               values=["Price", "Quantity"],
               columns=["Product"],
               aggfunc=[np.sum], fill_value=0)
```

```python
pd.pivot_table(data, index=["Manager","Rep"], 
               values=["Price", "Quantity"],
               columns=["Product"],
               aggfunc=[np.sum], fill_value=0,
               margins=True)
```

```python
# 중간정리) 내가 원하는대로 판을 짜겠다
#  가로: index
#  세로: values
#  무엇으로 채울까 : aggfunc  --> 패키지함수, 파이썬함수 etc (함수명)
#  세로를 항목별로 나눠서 보고 싶으면 : columns

# ----> 보고자 하는 항목 세로를 각기 다른 연산으로 대표하고 싶다
#       가격은 총합, 판매수량은 단순 카운틴  count()
#  ---> 각각 적용하고자 하는 것이 다를 때 dict
```

```python
pd.pivot_table(data, index=["Manager","Rep"], 
               values=["Price", "Quantity"],
               columns=["Product"],
               aggfunc={"Price":[np.sum, np.mean],
                       "Quantity":len},
               fill_value=0)
```

```python
# 이 중에서 내가 보고자 하는 조건을 다시 검색할 때...
# 소개) 저는 잘;;;;

table = pd.pivot_table(data, index=["Manager","Rep"], 
               values=["Price", "Quantity"],
               columns=["Product"],
               aggfunc={"Price":[np.sum, np.mean],
                       "Quantity":len},
               fill_value=0)
table.query('Manager=="Debra Henley"')
```

```python
#### 카운트로 채우는 crootable !!!!!
# 주어진 가로/세로의 조건에 맞는 데이터가 몇 건이 있는지!!!
# 단순 카운팅!!!!!! 
# crosstab : 채우는 방식 무조건 카운팅!!!!
```

```python
pd.crosstab(data["Name"], data["Price"])
```

```python
pd.crosstab([data["Manager"], data["Rep"]], [data["Price"], data["Quantity"]])
```
