### 리뷰 크롤링

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By 
```

```python
df["search_name"] = df["행정동명"] + " " + df["상호명"]
df["search_name"]
```

```python
df["url"] = ""
```

```python
df
```

```python
for i, name in enumerate(df['search_name'].tolist()):
    try:
        if __name__ == "__main__":
            options = Options()
            options.headless = True  # 브라우저 자동화 작업을 화면에 표시하지 않게 됨
            driver = webdriver.Chrome(executable_path=ChromeDriverManager().install(), options=options)
            url = f"https://m.map.naver.com/search2/search.naver?query={name}&sm=hty&style=v5"
            driver.get(url)
            df.iloc[i,-1] = driver.find_element(By.XPATH, "/html/body/div[4]/div[2]/ul/li/div[1]/a[2]").get_attribute("data-cid")

    except Exception as e1:
        continue


driver.quit()
```

```python
new_df = pd.merge(no_dupl[["상호명", "url"]], new_df, how="left", on="url")
```

```python
new_df["score"] = ""
new_df["review_cnt"] = ""
new_df["pos"] = ""
new_df["review"] = ""
new_df
```

```python
new_df["url"] = "https://m.place.naver.com/restaurant/" + new_df["url"]
new_df
```

```python

for i, url in enumerate(new_df["url"].tolist()):
    url1 = url
    url2 = url + '/review/visitor'
    if __name__ == "__main__":
        options = Options()
        options.headless = True
        driver = webdriver.Chrome(executable_path=ChromeDriverManager().install(), options=options)
        sub_driver = webdriver.Chrome(executable_path=ChromeDriverManager().install(), options=options)
        driver.get(url1)
        sub_driver.get(url2)
    score = 0
    review_cnt = 0
    pos = ""
    
    try :
        try:
            a = '/html/body/div[3]/div/div/div/div[2]/div[1]/div[2]/span[1]/em'
            score = driver.find_element(By.XPATH,a).text
        except:
            score = 0
            
        try:   
            b = '/html/body/div[3]/div/div/div/div[2]/div[1]/div[2]/span[2]/a/em'
            review_cnt = driver.find_element(By.XPATH,b).text
        except:
            review_cnt = 0 
            
        try:
            c = "/html/body/div[3]/div/div/div/div[6]/div/div[1]/div/ul/li[1]/div/div/strong"
            pos = driver.find_element(By.XPATH, c).text
        except:
            c = "/html/body/div[3]/div/div/div/div[6]/div/div[2]/div/ul/li[1]/div/div/strong"
            pos = driver.find_element(By.XPATH, c).text

        reviews = sub_driver.find_elements(By.CLASS_NAME, "YeINN")
        review = ""
        for j in range(len(reviews)):
            review += reviews[j].find_element(By.CLASS_NAME, "zPfVt").text + " "

        new_df.iat[i, 2] = score
        new_df.iat[i, 3] = review_cnt
        new_df.iat[i, 4] = pos
        new_df.iat[i, 5] = review
    except:
        new_df.iat[i, 2] = "N"
        new_df.iat[i, 3] = "N"
        new_df.iat[i, 4] = "N"
        new_df.iat[i, 5] = "N"
        print(f"{i}번째에서 에러 발생")

driver.quit()
```
